```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
import cv2
```

```python
path_images = './Images/withNoise/'
sigma_noise_background = 50
```

```python
#Function to get the sigma of the respective soure noise.
def get_sigma_source_noise(sigma_noise_background, amplitude_source):
    return np.sqrt(sigma_noise_background**2.0 + amplitude_source)
```

```python
#Generates a black image
width = 255
height = 255
blank_image = np.zeros((height,width), np.uint8)
cv2.imwrite(path_images+'blackImage.jpg', blank_image) 
plt.imshow(blank_image, cmap="Greys_r")
```

```python
canvas = cv2.imread(path_images+'blackImage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
cv2.circle(canvas, (width/2,height/2), 1, (255, 255, 255), -1) 
cv2.circle(canvas, (width/3, height/4), 1, (180, 180, 180), -1)
cv2.ellipse(canvas,(int(width*0.8), int(height*0.4)),(20,40),30,0,360,180,-1)
cv2.imwrite(path_images+'truesky_ref.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas = cv2.imread(path_images+'truesky_ref.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
cv2.circle(canvas, (64, 180), 1, (255, 255, 255), -1) # an asteroid !!
cv2.circle(canvas, (210, 120), 1, (255, 255, 255), -1) # a supernova !!! (no superman this time :( )
cv2.imwrite(path_images+'truesky_new.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas = cv2.imread(path_images+'blackImage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
row, col = canvas.shape
mean = 0
gauss = np.random.normal(mean, sigma_noise_background, (row,col))
gauss = gauss.reshape(row, col)
canvas = canvas + gauss
cv2.imwrite(path_images+'noisy_r.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas = cv2.imread(path_images+'blackImage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
row, col = canvas.shape
mean = 0
gauss = np.random.normal(mean, sigma_noise_background, (row,col))
gauss = gauss.reshape(row, col)
canvas = canvas + gauss
cv2.imwrite(path_images+'noisy_n.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas1 = cv2.imread(path_images+'noisy_r.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE) 
canvas2 = cv2.imread(path_images+'truesky_ref.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
canvas = cv2.GaussianBlur(canvas2, (3,3), 3, 3)
canvas = canvas + canvas1 
cv2.imwrite(path_images+'ReferenceImage.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas1 = cv2.imread(path_images+'noisy_n.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE) 
canvas2 = cv2.imread(path_images+'truesky_new.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
canvas = cv2.GaussianBlur(canvas2, (3,3), 3, 3)
canvas = canvas + canvas1 
cv2.imwrite(path_images+'NewImage.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

We simulated a certain number of sources. With certain amplitudes. Let's insert the source noise generated by each one. As this is simplify we only have two types of amplitudes

```python
#The two type of amplitudes inserted
amp_1 = 255
amp_2 = 180
#Number of each type in the reference
ref_1 = 1
ref_2 = 1 + np.pi * 20*40
#Number of each type in the new image
new_1 = 3
new_2 = 1 + np.pi * 20*40
```

```python
#Gets the sigmas for the sources noise
sigma_1 = get_sigma_source_noise(sigma_noise_background, amp_1)
sigma_2 = get_sigma_source_noise(sigma_noise_background, amp_2)
print (sigma_1, sigma_2)
```

Now we generate the convolutions and add them to the respective images

```python
canvas = cv2.imread(path_images+'blackImage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
row, col = canvas.shape
mean = 0
gauss = np.random.normal(mean, sigma_1, (row,col))
gauss = gauss.reshape(row, col)
#canvas = canvas + new_1 * gauss
canvas = canvas + gauss
cv2.imwrite(path_images+'source_noise_1_n.jpg', canvas)
```

```python
canvas = cv2.imread(path_images+'blackImage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
row, col = canvas.shape
mean = 0
gauss = np.random.normal(mean, sigma_2, (row,col))
gauss = gauss.reshape(row, col)
#canvas = canvas + new_2 * gauss
canvas = canvas + gauss
cv2.imwrite(path_images+'source_noise_2_n.jpg', canvas)
```

```python
canvas = cv2.imread(path_images+'blackImage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
row, col = canvas.shape
mean = 0
gauss = np.random.normal(mean, sigma_1, (row,col))
gauss = gauss.reshape(row, col)
#canvas = canvas + ref_1 * gauss
canvas = canvas + gauss
cv2.imwrite(path_images+'source_noise_1_r.jpg', canvas)
```

```python
canvas = cv2.imread(path_images+'blackImage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
row, col = canvas.shape
mean = 0
gauss = np.random.normal(mean, sigma_2, (row,col))
gauss = gauss.reshape(row, col)
#canvas = canvas + ref_2 * gauss
canvas = canvas + gauss
cv2.imwrite(path_images+'source_noise_2_r.jpg', canvas)
```

```python
canvas1 = cv2.imread(path_images+'source_noise_1_r.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE) 
canvas2 = cv2.imread(path_images+'source_noise_2_r.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE) 
canvas3 = cv2.imread(path_images+'ReferenceImage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
#canvas = cv2.GaussianBlur(canvas2, (3,3), 3, 3)
canvas = canvas1 + canvas2 + canvas3
cv2.imwrite(path_images+'ReferenceImage.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas1 = cv2.imread(path_images+'source_noise_1_n.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE) 
canvas2 = cv2.imread(path_images+'source_noise_2_n.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE) 
canvas3 = cv2.imread(path_images+'NewImage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
#canvas = cv2.GaussianBlur(canvas2, (3,3), 3, 3)
canvas = canvas1 + canvas2 + canvas3
cv2.imwrite(path_images+'NewImage.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```
