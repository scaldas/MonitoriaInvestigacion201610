```python
%pylab inline
import cv2
```

```python
#Generates a black image
width = 255
height = 255
blank_image = np.zeros((height,width), np.uint8)
cv2.imwrite('./Images/blackImage.jpg', blank_image) 
plt.imshow(blank_image, cmap="Greys_r")
```

```python
canvas = cv2.imread('./Images/blackImage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
cv2.circle(canvas, (width/2,height/2), 1, (255, 255, 255), -1) 
cv2.circle(canvas, (width/3, height/4), 1, (180, 180, 180), -1)
cv2.ellipse(canvas,(int(width*0.8), int(height*0.4)),(20,40),30,0,360,180,-1)
cv2.imwrite('./Images/truesky_ref.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas = cv2.imread('./Images/truesky_ref.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
cv2.circle(canvas, (64, 180), 1, (255, 255, 255), -1) # an asteroid !!
cv2.circle(canvas, (210, 120), 1, (255, 255, 255), -1) # a supernova !!! (no superman this time :( )
cv2.imwrite('./Images/truesky_new.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas = cv2.imread('./Images/blackImage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
row, col = canvas.shape
mean = 0
gauss = np.random.normal(mean, 50, (row,col))
gauss = gauss.reshape(row, col)
canvas = canvas + gauss
cv2.imwrite('./Images/noisy.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas1 = cv2.imread('./Images/noisy.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE) 
canvas2 = cv2.imread('./Images/truesky_ref.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
canvas = cv2.GaussianBlur(canvas2, (3,3), 3, 3)
canvas = canvas + canvas1 
cv2.imwrite('./Images/ReferenceImage.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas1 = cv2.imread('./Images/noisy.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE) 
canvas2 = cv2.imread('./Images/truesky_new.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
canvas = cv2.GaussianBlur(canvas2, (3,3), 3, 3)
canvas = canvas + canvas1 
cv2.imwrite('./Images/NewImage.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```
